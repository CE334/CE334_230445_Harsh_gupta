### Answer 11

```python
# 11. Inverse, rank, condition number, singular value decomposition and norm
#  Inverse
inv_a = np.linalg.inv(a)
print('inverse of a:')
print(inv_a)

# Rank
rank_a = np.linalg.matrix_rank(a)
print(f'\nrank= {rank_a}')

# Condition number
cond_a = np.linalg.cond(a)
print(f'condition number= {cond_a}\n')

# Singular value decomposition
u, s, vh = np.linalg.svd(a)
print('U=')
print(u)
print('\nSigma=')
print(np.diag(s))
print('\nVT=')
print(vh)

# Norm
norm_a = np.linalg.norm(a)
print(f'\nnorm= {norm_a}')
```

### Answer 13

```python
# 13. Derivative function with forward, backward, and central methods
def derivative(f, x, method='central', h=0.01):
    if method == 'forward':
        return (f(x + h) - f(x)) / h
    elif method == 'backward':
        return (f(x) - f(x - h)) / h
    elif method == 'central':
        return (f(x + h) - f(x - h)) / (2 * h)
    else:
        raise ValueError("Method not found")

# Finding derivative at x = 0.5
exp_der = derivative(np.exp, 0.5)
cos_der = derivative(np.cos, 0.5)
print(f"Exp derivative at 0.5: {exp_der}")
print(f"Cos derivative at 0.5: {cos_der}")
```
